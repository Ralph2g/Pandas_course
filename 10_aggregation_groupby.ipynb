```
import pandas as pd
import numpy as np
```

```
pd.options.display.float_format = '{:,.3f}'.format
```

```
import seaborn as sns
```

```
df = sns.load_dataset('tips')
df
```

```
df.describe(include='all')
```

```
# contamos los días que tiene nuestro df
df['day'].value_counts()
```

```
# Sacamos el promedio
df['day'].value_counts(normalize = True)*100
```

```
# Agrupamos por sexo y calculamos el promedio en los valores numericos
df.groupby('sex').mean()
```

```
# Calculamos el porcentaje de la propina en cada orden
df['prct_tip'] = df['tip']/df['total_bill']
df
```

```
# Vemos que porcentaje de propina recibe cada sexo
df.groupby('sex').mean()
```

```
# Por los valores de sengo en el promedio mejor sacamos la mediana
df.groupby('sex').median()
```

```
# Podemos agragar mas columnas descriptivas con la siguiente manera:
df.groupby('sex')['total_bill','prct_tip'].describe()
```

```
# Ahora definamos funciones que podemos implementar en nuestro datset con apply
# y también con aggregate
def mean_eur2usd(x):
  y = np.mean(x)*1.12
  return y
df.groupby('sex')[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```
# Se pueden agregar más columnas para aplicar estas funciones con apply
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```
# También podemos implmentar funciones de numpy

df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(np.std)
```

```
#También se pueden aplicar funciones lambda 
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(lambda x:np.mean(x)*1.12)
```

```
# Pero si queremos aplicar más de una funcion a nuestro df usamos aggregate
df.groupby(['sex','time'])[['total_bill','prct_tip']].aggregate([np.mean,np.max])
```

```
# Agregate mas corto
df.groupby(['sex','time'])[['total_bill','prct_tip']].agg([np.mean,np.max])
```

```
# podemos definir un diccionario de  funciones a aplicar a columnas especificas
dict_agg = {'tip':[min,max], 'total_bill':[np.mean,mean_eur2usd]}
dict_agg
```

```
df.groupby(['sex','time'])[['total_bill','tip']].agg(dict_agg)
```

```
# Se pueden aplicar filtros a las agrupaciones como se definira en la sig fun
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 20
```

```
# ahora se lo aplicamos al grupo de la siguiente manera
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
```

```
# Si ibservamos ahora tenemos menos elementos en nuestro df
# contemos los valores de cada columna y fila

df_filtered.groupby(['sex','time']).count()
```

### Segunda parte
## Extraer valores de variables categoricas

```
df
```

```
# Agregamos una columna de 1's que nos servira para contar
df['ones'] = 1
```

```
df
```

```
df_g = df.groupby(['sex','smoker'])[['ones']].sum()
df_g
```

```
# apliquemos el pocentaje de las personas que fuman con una función lambda

df_g.groupby(level=0).apply(
    lambda x:
    x/x.sum()*100
)
```

```
# También podemos crear categorias en nuestros df en este caso vamos a hacerlo de la siguinte manera
df['bin_total'] = pd.cut(df['total_bill'], bins=[3,18,35,60])
df
```

```
# contemos los valores que tenemos para cada categoria 
df.groupby(['time','bin_total'])[['ones']].count()
```

```
# Para hacerlo mucho más visible vamos a sacar los porcentajes con la funcion lambda
df.groupby(['time','bin_total'])[['ones']].count().groupby(level=0).apply(
    lambda x:
    x/x.sum()*100
)
```

## Ahora vamos a ver que hace pivot

### esto sirve para hacer tablas dinamicas

```
df
```

```
df.groupby(['sex','time'])[['total_bill']].mean()
```

```
df_gp = df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
df_gp
```

```
df_gp.pivot_table(values='total_bill',index='sex',columns='time')
```

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

```
# Como vemos esta tabla la podemos sacar directamente del dataframe
# Algo interesante es de que esta a su vez ejecuta una función por defecto
# función que se puede editar en la función group_by

df.pivot_table(values='total_bill',index='sex',columns='time')
```

```
df.pivot_table(values='total_bill',index='sex',columns='time',aggfunc=[np.median])
```

```
# También se puede crear un diccionario de funciones y a su vez también se puede 
# anidar las funciones
df_pivot = df.pivot_table(values='total_bill',index='sex',columns='time',aggfunc=[np.median,np.std])
df_pivot
```

```
# Ahora si queremos deshacer esta tabla lo que necesitamos hacer es llamar a la función 
# unstack
df_pivot.unstack().reset_index()
```
